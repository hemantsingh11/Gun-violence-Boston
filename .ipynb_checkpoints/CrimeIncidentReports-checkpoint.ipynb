```python
# filtering city council diistrict 4 data
cir_D4_df = cir_df[cir_df['DISTRICT'] == 'D4']
```

```python
# removing rows which didnt have coordinates
cir_df = cir_df.dropna(subset=['Lat', 'Long', 'Location'])
```

```python
# removing rows which didnt have coordinates
cir_D4_df = cir_D4_df.dropna(subset=['Lat', 'Long', 'Location'])
```

```python
# showing count of each offense description
grouped = cir_df.groupby(['OFFENSE_DESCRIPTION']).size().reset_index(name='COUNT').sort_values('COUNT', ascending=False).reset_index(drop=True)
grouped['Index'] = grouped.index + 1
#grouped.head(60)
grouped['OFFENSE_DESCRIPTION'][45]
```

```python
# top 40 offense description count
countGT100_offenseDesc = grouped.head(40)
```

```python
# top 10 offense description count
top10_offense_desc = ['ASSAULT - SIMPLE', 'PROPERTY - LOST/ MISSING','LARCENY SHOPLIFTING','VERBAL DISPUTE',
                      'LARCENY THEFT FROM MV - NON-ACCESSORY','LARCENY THEFT FROM BUILDING','ASSAULT - AGGRAVATED','THREATS TO DO BODILY HARM','DRUGS - POSSESSION/ SALE/ MANUFACTURING/ USE','LARCENY ALL OTHERS']

# offense description count greater than 2000
countGT2k_offense_desc = ['ASSAULT - SIMPLE', 'PROPERTY - LOST/ MISSING','LARCENY SHOPLIFTING']

# offense description count between 1000 and 2000
countGT1k_offense_desc = ['VERBAL DISPUTE','LARCENY THEFT FROM MV - NON-ACCESSORY','LARCENY THEFT FROM BUILDING',
                          'ASSAULT - AGGRAVATED','THREATS TO DO BODILY HARM','DRUGS - POSSESSION/ SALE/ MANUFACTURING/ USE',
                          'LARCENY ALL OTHERS', 'HARASSMENT/ CRIMINAL HARASSMENT','VAL - VIOLATION OF AUTO LAW',
                         'MISSING PERSON - LOCATED']
```

```python
# map of city council district border and crime incident 
m = folium.Map(location=[cir_df['Lat'].mean(), cir_df['Long'].mean()], zoom_start=12)

# add a layer for crime incident report markers
cir_layer = folium.FeatureGroup(name='Crime Incident Reports')
for index, row in cir_df.iterrows():
    location = [row['Lat'], row['Long']]
    tooltip = f"{row['OFFENSE_DESCRIPTION']}, {row['DISTRICT']}"
    radius = 5  # size of data point
    offense = row['OFFENSE_DESCRIPTION']
    if offense in countGT100_offenseDesc['OFFENSE_DESCRIPTION'].tolist():
        color = 'red'
    else:
        color = 'blue'
    fill_color = color  # fill color same as border color
    fill_opacity = 0.5  # opacity of fill color
    folium.Circle(location=location, tooltip=tooltip, radius=radius,
                  color=color, fill_color=fill_color, fill_opacity=fill_opacity).add_to(cir_layer)
cir_layer.add_to(m)

# add a layer for city council districts GeoJSON file
geojson_data = open('City_Council_Districts_-_Effective_for_the_2023_Municipal_Election.geojson')
cc_layer = folium.FeatureGroup(name='City Council Districts')
folium.GeoJson(geojson_data.read()).add_to(cc_layer)
cc_layer.add_to(m)

# add layer control to switch between layers
folium.LayerControl().add_to(m)

# create a custom legend for the circle colors
legend_html = '''
<div style="position: fixed; 
     bottom: 50px; left: 50px; width: 150px; height: 90px; 
     border:2px solid grey; z-index:9999; font-size:14px;
     background-color: white;
     ">
     &nbsp; Circle Colors <br>
     &nbsp; Offense count>100; <i class="fa fa-circle fa-1x" style="color:red"></i><br>
     &nbsp; Other Offenses &nbsp; <i class="fa fa-circle fa-1x" style="color:green"></i>
</div>
'''

# add the legend to the map
m.get_root().html.add_child(folium.Element(legend_html))

# save map as an HTML file
m.save('CrimeIncidentReport.html')
```

```python
larceny_theft = ['AUTO THEFT', 'ROBBERY', 'BURGLARY - RESIDENTIAL', 'BURGLARY - COMMERICAL', 
                 'AUTO THEFT - MOTORCYCLE / SCOOTER',
                 'LARCENY PICK-POCKET', 'LARCENY PURSE SNATCH - NO FORCE', 
                 'LARCENY THEFT FROM COIN-OP MACHINE', 'POSSESSION OF BURGLARIOUS TOOLS',]


drug_related= ['DRUGS - POSSESSION/ SALE/ MANUFACTURING/ USE', 
               'DRUGS - POSSESSION OF DRUG PARAPHANALIA', 
               'OPERATING UNDER THE INFLUENCE (OUI) DRUGS']

murder= ['MURDER, NON-NEGLIGENT MANSLAUGHTER', 
         'MURDER, NON-NEGLIGIENT MANSLAUGHTER', 'MANSLAUGHTER - VEHICLE - NEGLIGENCE']


gun_related= ['BALLISTICS EVIDENCE/FOUND', 'FIREARM/WEAPON - FOUND OR CONFISCATED', 
               'BOMB THREAT', 'FIREARM/WEAPON - LOST', 'FIREARM/WEAPON - ACCIDENTAL INJURY / DEATH',
               'EXPLOSIVES - POSSESSION OR USE', 'EXPLOSIVES - TURNED IN OR FOUND', 
               'WEAPON VIOLATION - CARRY/ POSSESSING/ SALE/ TRAFFICKING/ OTHER']

fraud_related= ['FRAUD - FALSE PRETENSE / SCHEME', 'FORGERY / COUNTERFEITING', 'FRAUD - WIRE', 'FRAUD - IMPERSONATION']

kidnaping = ['KIDNAPPING/CUSTODIAL KIDNAPPING/ ABDUCTION']

fugitive = ['FUGITIVE FROM JUSTICE', 'PRISONER ESCAPE / ESCAPE & RECAPTURE']

warrant_arrest = ['WARRANT ARREST - BOSTON WARRANT (MUST BE SUPPLEMENTAL)', 'WARRANT ARREST - OUTSIDE OF BOSTON WARRANT	']
```

```python
# map of city council district border and crime incident 
m = folium.Map(location=[cir_df['Lat'].mean(), cir_df['Long'].mean()], zoom_start=12)

# add a layer for crime incident report markers
cir_layer = folium.FeatureGroup(name='Crime Incident Reports')
for index, row in cir_df.iterrows():
    location = [row['Lat'], row['Long']]
    tooltip = f"{row['OFFENSE_DESCRIPTION']}, {row['DISTRICT']}"
    radius = 5  # size of data point
    offense = row['OFFENSE_DESCRIPTION']
    if offense in larceny_theft:
        color = 'Yellow'
    if offense in drug_related:
        color = 'Blue'
    if offense in murder:
        color = 'Blue'
    if offense in gun_related:
        color = 'red'
    if offense in fraud_related:
        color = 'Green'
    if offense in kidnaping:
        color = 'red'
    if offense in fugitive:
        color = 'Orange'
    if offense in warrant_arrest:
        color = 'Black'
    else:
        color = 'Gray'
    fill_color = color  # fill color same as border color
    fill_opacity = 0.5  # opacity of fill color
    folium.Circle(location=location, tooltip=tooltip, radius=radius,
                  color=color, fill_color=fill_color, fill_opacity=fill_opacity).add_to(cir_layer)
cir_layer.add_to(m)

# add a layer for city council districts GeoJSON file
geojson_data = open('City_Council_Districts_-_Effective_for_the_2023_Municipal_Election.geojson')
cc_layer = folium.FeatureGroup(name='City Council Districts')
folium.GeoJson(geojson_data.read()).add_to(cc_layer)
cc_layer.add_to(m)

# add layer control to switch between layers
folium.LayerControl().add_to(m)

# create a custom legend for the circle colors
legend_html = '''
<div style="position: fixed; 
     bottom: 50px; left: 50px; width: 150px; height: 220px; 
     border:2px solid grey; z-index:9999; font-size:14px;
     background-color: white;
     ">
     &nbsp; Circle Colors <br>
     &nbsp; Larceny/Theft &nbsp; <i class="fa fa-circle fa-1x" style="color:yellow"></i><br>
     &nbsp; Drug Related &nbsp; <i class="fa fa-circle fa-1x" style="color:blue"></i><br>
     &nbsp; Murder &nbsp; <i class="fa fa-circle fa-1x" style="color:blue"></i><br>
     &nbsp; Gun Related &nbsp; <i class="fa fa-circle fa-1x" style="color:red"></i><br>
     &nbsp; Fraud Related &nbsp; <i class="fa fa-circle fa-1x" style="color:green"></i><br>
     &nbsp; Kidnaping &nbsp; <i class="fa fa-circle fa-1x" style="color:red"></i><br>
     &nbsp; Fugitive &nbsp; <i class="fa fa-circle fa-1x" style="color:orange"></i><br>
     &nbsp; Warrant Arrest &nbsp; <i class="fa fa-circle fa-1x" style="color:black"></i><br>
     &nbsp; Other Offenses &nbsp; <i class="fa fa-circle fa-1x" style="color:gray"></i>
</div>
'''

# add the legend to the map
m.get_root().html.add_child(folium.Element(legend_html))

# save map as an HTML file
m.save('CrimeIncidentReport.html')
```
